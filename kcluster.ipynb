```
#k 
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

from sklearn.preprocessing import scale
from sklearn.preprocessing import normalize

#model
from sklearn.cluster import KMeans


```

```
data =pd.read_excel("university.xlsx")
data.head()
```

```
data.isnull().sum()
```

```
data1=data.iloc[:,1:]
data1.head()
```

```
data1.shape
```

```
data_new =scale(data1)
data_new
```

```
data2_new= normalize(data1)
data2_new
```

```
#model
model= KMeans(n_clusters=3)
```

```
model.fit(data2_new)
```

```
model.labels_
```

```
# convert 1D to 2D
cluster = pd.DataFrame(model.labels_)
```

```
cluster.head()
```

```
data['Tire']= cluster
```

```
data.head()
```

```
data1.to_excel("university.xlsx", sheet_name='sheet2' )
```
