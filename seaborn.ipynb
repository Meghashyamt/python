```
#Distribution
## Distplot
## Joint plot
## Pair plot
## Rug plot
## Kdeplot
```

```
import seaborn as sns
%matplotlib inline
```

```
tips =sns.load_dataset('tips')
```

```
tips.head()
```

```
sns.distplot(tips['total_bill'], kde= False, bins=30)
#bins number of distribution
#kde removes line curve
```

```
sns.distplot(tips['total_bill'])
```

```
iris = sns.load_dataset('titanic')
iris.head()
```

```
sns.countplot(x='embarked', data=iris)
```

```
sns.pairplot(iris)
```

```
#jointplot
sns.jointplot(x='total_bill' , y= 'tip', data= tips, kind='reg')
```

```
#pair plot
#hue categroical columns
sns.pairplot(tips ,hue='day')
```

```
#Rug plot
sns.rugplot(tips['total_bill'])
```

```
#kde plot
sns.kdeplot(tips['total_bill'])
```

```
sns.rugplot(tips['total_bill'])
sns.kdeplot(tips['total_bill'])
```

```
#catgeorical data plots
## box plot
## violin plot

## stripplot
## swarmplot

## bar plot
## count plot

## factor plot
```

```
#bar plot
sns.barplot(x='sex', y='total_bill', data= tips)
```

```
import numpy as np
sns.barplot(x='sex', y='total_bill', data= tips, estimator= np.std)
```

```
#  count plot
sns.countplot(x='sex', data=tips)
```

```

sns.countplot(x='day', data = tips)
```

```
#box plot
sns.boxplot(x='sex', y='total_bill', data = tips,palette='rainbow')
```

```
sns.boxplot(x='day', y='total_bill', data = tips,palette='rainbow')
```

```
sns.boxplot(x='total_bill', y='sex', data = tips,palette='rainbow')
```

```
sns.boxplot(x='pclass', y='fare', data = iris,palette='rainbow')
```

```
sns.boxplot(data = tips,palette='rainbow', orient= 'h')
```

```
sns.boxplot(x='sex', y='total_bill', data = tips,palette='rainbow', hue ='smoker')
```

```
#violen plot
sns.violinplot(x='sex', y='total_bill', data = tips,palette='rainbow')
```

```
sns.violinplot(x='sex', y='total_bill', data = tips,palette='rainbow', hue='smoker', split= True)
```

```
# strip plot
sns.stripplot(x='day', y='total_bill',data=tips, size=5)
```

```
#swam plot
sns.swarmplot(x='day', y='total_bill',data=tips, size=5, hue='sex', split= True)
```

```
#factor plot
sns.factorplot(x='day', y='total_bill',data=tips, size=5)
```
